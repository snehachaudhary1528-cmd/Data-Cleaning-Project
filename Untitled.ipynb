```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('dataset.xlsx')
```

```python
print("=" * 50)
print("ORIGINAL DATASET")
print("=" * 50)
print(df.head(10)) 
print("=" * 50)
print(f"\nTotal Rows: {len(df)}, Total Columns: {len(df.columns)}")
print("=" * 50)
```

```python
print("\n" + "=" * 50)
print("DATASET STRUCTURE")
print("=" * 50)

print("\nColumns aur unke Data Types:")
print(df.dtypes)

print("\nBasic Statistics (Numeric columns):")
print(df.describe())
```

```python
print("\n" + "=" * 50)
print("MISSING VALUES")
print("=" * 50)

print("\nHar column mein missing values:")
print(df.isnull().sum())

print("\nMissing values ka percentage:")
print(round((df.isnull().sum() / len(df)) * 100, 2))
```

```python
print("\n" + "=" * 50)
print("DUPLICATES")
print("=" * 50)

duplicates = df.duplicated().sum()
print(f"\nTotal duplicate rows: {duplicates}")

if duplicates > 0:
    print("\nDuplicate rows (first 5):")
    print(df[df.duplicated(keep=False)].head())
```

```python
print("\n" + "=" * 50)
print("CLEANING STARTED...")
print("=" * 50)
```

```python
df_clean = df.drop_duplicates()
print(f"\n1️ Duplicates cleared: {len(df)} → {len(df_clean)} rows")
```

```python
num_cols = df_clean.select_dtypes(include=[np.number]).columns
df_clean[num_cols] = df_clean[num_cols].fillna(0)
print(f"2️ Numeric columns ({list(num_cols)}) are filled with 0")
```

```python
text_cols = df_clean.select_dtypes(include=['object']).columns
df_clean[text_cols] = df_clean[text_cols].fillna('Unknown')
print(f"Text columns ({list(text_cols)}) are filled with unknown")
```

```python
bool_cols = df_clean.select_dtypes(include=['bool']).columns
if len(bool_cols) > 0:
    df_clean[bool_cols] = df_clean[bool_cols].fillna(False)
    print(f" Boolean columns ({list(bool_cols)}) missing values are filled with False")
```

```python
print("\n" + "=" * 50)
print("FINAL QUALITY CHECK")
print("=" * 50)


print(df_clean.isnull().sum().sum())  

print(df_clean.duplicated().sum()) 
```

```python
df_clean.to_excel('cleaned_dataset.xlsx', index=False)
```

```python
print("\n" + "=" * 50)
print("SUMMARY")
print("=" * 50)
print(f"\n Original file: dataset.xlsx ({len(df)} rows)")
print(f" Cleaned file: cleaned_dataset.xlsx ({len(df_clean)} rows)")
print(f" Duplicates removed: {len(df) - len(df_clean)}")
print(f" Missing values filled: {df.isnull().sum().sum() - df_clean.isnull().sum().sum()}")
print("\n Data Cleaning Complete! 'cleaned_dataset.xlsx' saved")
```
